```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Store the datetimes when the weight measurements were taken
dates_and_weights_dict = {
    "2024-03-22": 3.520,
    "2024-03-27": 3.260,
    "2024-04-01": 3.130,
    "2024-04-02": 3.100,
    "2024-04-03": 3.200,
    "2024-04-06": 3.220,
    "2024-04-08": 3.380,
}

# Convert to a DataFrame for easier plotting
data_df = pd.DataFrame(list(dates_and_weights_dict.items()), columns=["Date", "Weight"])
data_df["Date"] = pd.to_datetime(data_df["Date"])
data_df.sort_values("Date", inplace=True)
data_df.head()
```

```python
# Plot the weights over time
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Weight", data=data_df, marker="o")
plt.title("Charlie's Weight Over Time")
plt.xlabel("Date")
plt.ylabel("Weight (kg)")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()

# Show the xtick labels for the dates of measurements
plt.xticks(data_df["Date"], rotation=45)
plt.show()
```
